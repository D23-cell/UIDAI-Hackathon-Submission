```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

print("Libraries Imported Successfully.")

def load_and_combine_csvs(folder_name):
    """
    Reads all CSV files from a specific folder and combines them into a single DataFrame.
    """
    path = f"/content/{folder_name}"
    csv_files = glob.glob(f"{path}/*.csv")

    if not csv_files:
        print(f"‚ö†Ô∏è Warning: No CSV files found in '{folder_name}'")
        return None

    print(f"üìÇ Processing '{folder_name}' ({len(csv_files)} files found)...")

    df_list = []
    for file in csv_files:
        try:
            df = pd.read_csv(file, low_memory=False)
            df_list.append(df)
        except Exception as e:
            print(f"‚ùå Error reading {os.path.basename(file)}: {e}")

    if df_list:
        combined_df = pd.concat(df_list, ignore_index=True)
        print(f"‚úÖ Successfully loaded '{folder_name}' with {len(combined_df)} rows.")
        return combined_df
    return None

print("\n--- Loading Datasets ---")
df_bio = load_and_combine_csvs("api_data_aadhar_biometric")
df_update = load_and_combine_csvs("api_data_aadhar_demographic")
df_enroll = load_and_combine_csvs("api_data_aadhar_enrolment")
```

```python

print("Processing Data for Geospatial Analysis...")

def preprocess_dates(df):
    if df is not None and 'date' in df.columns:
        df['date'] = pd.to_datetime(df['date'], errors='coerce')
        df['YearMonth'] = df['date'].dt.to_period('M')
    return df

df_bio = preprocess_dates(df_bio)
df_update = preprocess_dates(df_update)

def aggregate_by_district(df, label):
    """
    Aggregates update counts by District across age groups.
    """
    if df is None: return None

    age_cols = [c for c in df.columns if 'age' in c]

    df['Total_Updates'] = df[age_cols].sum(axis=1)

    district_group = df.groupby('district')['Total_Updates'].sum().reset_index()
    district_group.columns = ['District', f'{label}_Count']
    return district_group

dist_bio = aggregate_by_district(df_bio, 'Biometric')
dist_demo = aggregate_by_district(df_update, 'Demographic')

analysis_df = None
if dist_bio is not None and dist_demo is not None:
    analysis_df = pd.merge(dist_bio, dist_demo, on='District', how='outer').fillna(0)
    analysis_df['Total_Activity'] = analysis_df['Biometric_Count'] + analysis_df['Demographic_Count']

    top_districts = analysis_df.sort_values(by='Total_Activity', ascending=False).head(10)

    print("\n‚úÖ Data Aggregation Complete.")
    print("Top 5 Districts by Activity Volume:")
    print(top_districts[['District', 'Total_Activity']].head())
else:
    print("‚ùå Error: Insufficient data for district analysis.")
```

```python

if analysis_df is not None:

    top_10 = analysis_df.sort_values(by='Total_Activity', ascending=False).head(10)

    plt.figure(figsize=(14, 7))

    p1 = plt.bar(top_10['District'], top_10['Biometric_Count'], label='Biometric Updates', color='#d32f2f', alpha=0.9)
    p2 = plt.bar(top_10['District'], top_10['Demographic_Count'], bottom=top_10['Biometric_Count'], label='Demographic Updates', color='#1976d2', alpha=0.9)

    plt.title('Operational Stress Zones: Biometric vs Demographic Update Volume', fontsize=16, fontweight='bold')
    plt.xlabel('District', fontsize=12)
    plt.ylabel('Volume of Transactions', fontsize=12)
    plt.legend()
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    plt.savefig('District_Load_Analysis.png', bbox_inches='tight', dpi=300)
    plt.show()
    print("üìä Graph 1 Generated: District_Load_Analysis.png")

    if df_bio is not None:
        kids_col = [c for c in df_bio.columns if '5_17' in c][0]
        adult_col = [c for c in df_bio.columns if '17_' in c][0]

        kids_updates = df_bio[kids_col].sum()
        adult_updates = df_bio[adult_col].sum()

        plt.figure(figsize=(8, 8))
        colors = ['#ff9800', '#4caf50']
        explode = (0.05, 0)

        plt.pie([kids_updates, adult_updates], labels=['Children (5-17)', 'Adults (18+)'],
                autopct='%1.1f%%', colors=colors, startangle=90, explode=explode, shadow=True)

        plt.title('Biometric Update Distribution by Age Group', fontsize=16, fontweight='bold')

        plt.savefig('Age_Demographic_Split.png', dpi=300)
        plt.show()
        print("üìä Graph 2 Generated: Age_Demographic_Split.png")
```

```python

if analysis_df is not None:

    threshold = analysis_df['Biometric_Count'].quantile(0.75)

    critical_districts = analysis_df[analysis_df['Biometric_Count'] > threshold].copy()
    critical_districts['Recommendation'] = "Deploy Mobile Biometric Kits"

    critical_districts = critical_districts.sort_values(by='Biometric_Count', ascending=False)

    print(f"\nüöÄ INSIGHT: Identified {len(critical_districts)} Critical Districts requiring intervention.")

    filename = 'Critical_District_Action_Plan.csv'
    critical_districts.to_csv(filename, index=False)
    print(f"‚úÖ Solution File Saved: {filename}")

    print("\nSample of Critical Districts:")
    print(critical_districts[['District', 'Biometric_Count', 'Recommendation']].head())
```
